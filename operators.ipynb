### Arithmetic operators

```python
num1 = 12
num2 = 23
print(num1+num2) # add
print(num1-num2) # sub
print(num1*num2) # multi
print(num1/num2) # division ---> decimal (float)
print(num1//num2) # floor division ---> int
print(num1**num2) # expo 
```

### comparision (relational)

```python
num3 = 10
num4 = 10
num5 = 12
print(num3 == num5) # equal to
print(num3 != num5) # not equal to
print(num3 > num4) # greater then
print(num3 >= num4) # greater  then equal

print(num3 < num5) # less then
print(num3 <= num5) # less then equal to
```

#### membership

```python
text = "this is 10am batch"
print("is" in text)
print("iss" not in text)
```

### logical --> and,or,not

```python
tea = True
coffee = False
print(tea and coffee)
print(tea or coffee)
```

```python
print(True or False)
```

```python
print(not True)
print(not False)
```

#### identity ----> memory location

```python
num10 = 10
num11 = 10
num12 = 12
print(id(num10))
print(id(num11))
print(id(num12))
print(num10 is num11)
print(num10 is num12)
```
