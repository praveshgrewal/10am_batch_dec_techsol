### data types

##### 1. int ---> number

```python
age = 23 # int
print(age)
print(type(age)) # type method for check the type of data 
```

#### 2. float --> decimal

```python
score = 45.7
print(score)
print(type(score))
```

#### string --> text

#### boolean --> True/False

```python
t = True
print(t)
print(type(t))
f = False
print(f)
print(type(f))
```

#### None

```python
no = None
print(no)
print(type(no))
```

#### complex number (real + img number)

```python
com = 3+5j
print(com)
print(type(com))
```

#### list ---> [ ]

```python
my_list = [1,2,3,'hello',True]
print(my_list)
print(type(my_list))
```

#### tuple ---> ( )


```python
my_tuple =  (1,2,3,'hello',True)
print(my_tuple)
print(type(my_tuple))
```

#### set ---> { }

```python
my_set = {1,2,3,4,'''hello'''}
print(my_set)
print(type(my_set))
```

#### dictionary --> {'key':'value'}

```python
my_dict= {'name':'shaad','age':23,'address':'okhla vihar'}
print(my_dict)
print(type(my_dict))
```

#### type conversion
* implicit ---> python innerly
* explicit ---> we will do it manualy

```python
num1 = 10
print(type(num1))
num2 = 34.8
print(type(num2))

result = num1 + num2
print(result)
print(type(result))
```

```python
num4 = "shaad"
print(type(num4))
new_num = tuple(num4)
print(new_num)
print(type(new_num))
```
